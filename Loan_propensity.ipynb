```python



# Create PCTILE_score segmentation with values 1 /low probability to pay/,2 /average probablity to pay/ and 3 /high probability to pay/
# for given targets ap090, ct090
# for percentile score use all inputed values for ap090 and ct090
```

```python
# replace 1 column name by ap090
```

```python
rating_pctile = np.percentile(df7['ap090'], [75, 90])
```

```python
rating_pctile
```

```python
df7['PCTILE_score'] = 0
df7['PCTILE_score'] = np.where(df7['ap090'] < rating_pctile[0], 1, df7['PCTILE_score'])
df7['PCTILE_score'] = np.where((df7['ap090'] >= rating_pctile[0]) & (df7['ap090'] <= rating_pctile[1]), 2, df7['PCTILE_score'])
df7['PCTILE_score'] = np.where(df7['ap090'] > rating_pctile[1], 3, df7['PCTILE_score'])
```

```python
X = df7.drop(['ap090', 'PCTILE_score'], axis = 1)
y = df7['PCTILE_score']
```
